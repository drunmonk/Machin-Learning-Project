```python

# Load data
data1 = pd.read_csv("Global_Earthquake_Data.csv")


# Convert magnitude values to integer by rounding
data1['mag'] = data1['mag'].round()

# Create bins based on magnitude values
bins = [0, 4, 6, 100]
labels = ['low', 'mid', 'high']
data1['mag_bin'] = pd.cut(data1['mag'], bins=bins, labels=labels)

# Get count of each class
counts = data1['mag_bin'].value_counts()
# Plot count of each class
plt.bar(counts.index, counts.values)
plt.title("Count of Earthquakes by Magnitude Class")
plt.xlabel("Magnitude Class")
plt.ylabel("Count")
plt.show()
print(counts)
# Assuming your dataframe is called "df"
# Create a new column called "mag_type"
data1['mag_type'] = ""

# Fill the new column based on the "mag_bin" column values
data1.loc[data1['mag_bin'] == 'low', 'mag_type'] = 'L'
data1.loc[data1['mag_bin'] == 'mid', 'mag_type'] = 'M'
data1.loc[data1['mag_bin'] == 'high', 'mag_type'] = 'H'
data1 = data1[['latitude', 'longitude', 'mag', 'mag_type']]

# Print the first 10 rows of the resulting dataframe
print(data1.head(10))
```
